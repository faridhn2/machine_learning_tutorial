```python
weather = pd.read_csv('Temp_and_rain.csv')
```

```python
weather
```

```python
weather['Year'].value_counts()
```

```python
weather[weather['Year'] > 2013]
```

```python
weather['Month'] -= 1
```

```python
weather['Month']//3
```

```python
def season_saz(x):
  seasons = ['Bahar','Tabestan','Paeez','Zemestan']
  return seasons[x]
```

```python
weather['seasen'] = (weather['Month']//3).apply(season_saz)
```

```python
weather
```

```python
weather[weather['Year'] > 2013].pivot_table(values = 'rain' , index = ['Year' , 'seasen'] )
```

```python
weather['tem'] = weather['tem'].apply(int)
```

```python
weather['tem'].value_counts()
```

```python
weather
```

```python
weather[weather['Year'] > 2013].pivot_table(values = ['rain','tem' ], index = ['Year' , 'seasen'],aggfunc=['count','mean','max','min'])
```
