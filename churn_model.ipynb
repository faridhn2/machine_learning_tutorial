# Normalization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop("IsActiveMember", axis=1))
```

```python
feat_scaled = scaler.transform(df.drop("IsActiveMember", axis=1))
```

```python
feat_scaled
```

```python
col = list(df.columns)
col.remove('IsActiveMember')
col
```

```python
pd.DataFrame(feat_scaled,columns=col).isna().sum()
```

```python
X = pd.DataFrame(feat_scaled,columns=col)
```

```python
X.describe()
```

# Train & Test Split

```python
from sklearn.model_selection import train_test_split
# X = df.drop('IsActiveMember',axis=1)
y = df['IsActiveMember']
X_train , X_test , y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
X_train.isna().sum()
```

# Logestic Regression


```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, y_train)
log_pred = logistic_regression.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, log_pred))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=32)
knn.fit(X_train, y_train)
```

```python
knn_preds = knn.predict(X_test)
print(classification_report(y_test, knn_preds, output_dict=True)["accuracy"])
```

```python
import numpy as np

acc_rate = []
for k in range(1, 40):
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  knn_preds = knn.predict(X_test)
  acc_rate.append(classification_report(y_test, knn_preds, output_dict=True)["accuracy"])

print(np.argmax(acc_rate))
# print(classification_report(y_test, knn_preds, output_dict=True)["accuracy"])
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
dtree_preds = dtree.predict(X_test)
print(classification_report(y_test,dtree_preds))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators= 100)
rfc.fit(X_train,y_train)
rfc_preds = rfc.predict(X_test)
print(classification_report(y_test,rfc_preds))
```

# SVM

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,y_train)
svc_preds = svc.predict(X_test)
print(classification_report(y_test,svc_preds))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1,1, 10, 100, 1000],
              'gamma': [1,0.1,0.01,0.001,0.0001],
              'kernel': ['rbf']}
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=3)
grid.fit(X_train,y_train)
grid_preds = grid.predict(X_test)
print(classification_report(y_test, grid_preds))
```

# Results



*   Barplot (compare Acc in all models)




```python
outputs = [log_pred,knn_preds,dtree_preds,rfc_preds,svc_preds,grid_preds]
names = ['Log Reg.','KNN','DTree','RFC','SVM','GridCV']
accs = []
for output in outputs :
  acc = classification_report(y_test, output, output_dict=True)["accuracy"]
  accs.append(acc)

```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(names,accs,width=0.6);
plt.grid()
```
