```python
import re
```

```python
re.findall("\d+\.\d+",'19.34 Billion')[0]
```

```python
def convert_text(x):
  if not isinstance(x,float):

    if 'Trillion' in x:
      rate = 1000_000
    elif  'Billion' in x :
      rate = 1000
    else:
      rate = 1
    if '.' in x:
      result = re.findall("\d+\.\d+",x)[0]
    else:
      result = re.findall("\d+",x)[0]
    return float(result)*rate
  else:
    return 0
```

```python
df['GDP $USD'].isna().sum()
```

```python
df['GDP float M$'] = df['GDP $USD'].apply(convert_text)
```

```python
df[df['POPULATION']>df.describe()['POPULATION']['mean']].sort_values(by='POPULATION',ascending=False).reset_index()
```

```python
df['GDP/POP'] = df['GDP float M$']/df['POPULATION']
```
